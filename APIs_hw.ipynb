#### Types of APIs:

#####    Public APIs :
        These APIs are open and available for use by any outside developer or business to enable sharing of applications.
#####    Partner APIs:
        These APIs are suitable for  specific third-party business partner developers or API consumcers
#####    Composite APIs: 
        In this two or more APIs are combined for related operations to import speed and performance
#####     Internal APIs: 
        These APIs are used by internal enterprise to connect systems and data with the business

#### Types of API Architectures:

##### REST: 
    The majority of web APIs are built on REST. REST, which stands for representational state transfer. There are 6 rules that must be followed for an API to comply with REST (Representational State Transfer). 1. Cannot keep the information received from client/sender. 2. The client and server are independent of each other. Changes made on one doesn't affect the other. 3. Must have caching ability which allows information to be stored for a certain period in order to make the app more efficient and faster. 4. Have a uniform interface for communication. 5. Be a layered system, letting legacy code be encapsulated in one part of the API. 6. Adhere to code on demand allowing code to be transmitted within an application through the API.
##### RPC: 
    The RPC (Remote Procedural Call) protocol is the most straightforward of the three architectures. Unlike REST and SOAP that facilitate the transfer of data, RPC APIs invoke processes. This makes it easier for developers to create applications involving multiple programs or services.
##### SOAP: 
    Simple Object Access Protocol is a protocol for transmitting data across networks and can be used to build APIs. It's the offical protocol defined by the World Wide Web Consortium (W3C) and comes with strict rules.

#### Why are APIs important?
    APIs are a way for systems/applications to communicate and pass data. A well-written API can provide some very important long- term benefits. The importance of APIs from a technical standpoint, they allow the capabilities of one computer program to be used by another. They are a means by which two different programs are able to communicate.
